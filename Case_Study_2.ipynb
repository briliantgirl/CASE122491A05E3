```python
import pandas as cr
import matplotlib.pyplot as lr
import seaborn as cse

ntr = cr .read_csv('heart.csv')
print(ntr)

#scatterplot
cse.scatterplot(x='age',y='bp',data=ntr,hue='gender')
lr.title('Heart attack',color='red')
lr.xlabel('Age',color='green')
lr.ylabel('Bp',color='hotpink')
lr.show()

#linechart
cse.lineplot(x='age',y='cholestrol',data=ntr)
lr.show()

color=['pink','green']
cse.boxplot(x='target',y='value',data=ntr,palette=color)
lr.show()

color=['yellow','red']
cse.countplot(x='target',data=ntr,palette=color)
lr.show()

```
